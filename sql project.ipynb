```python
from pandasql import sqldf
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('ChicagoCrimeData .csv')

# SQL-like query to select unique crime types recorded at schools
query = """
    SELECT DISTINCT primary_type
    FROM df
    WHERE LOWER(location_description) LIKE '%school%';
"""

# Execute the query
unique_crime_types_at_schools = sqldf(query)

# Display the result
print(unique_crime_types_at_schools)
```

```python
from pandasql import sqldf
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('ChicagoPublicSchools.csv')

# SQL-like query to calculate the average safety score for each school ID
query = """
    SELECT school_id, AVG(safety_score) AS avg_safety_score
    FROM df
    GROUP BY school_id;
"""

# Execute the query
average_safety_score_by_id = sqldf(query)

# Display the result
print(average_safety_score_by_id)
```

```python
from pandasql import sqldf
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('ChicagoCensusData .csv')

# SQL-like query to calculate the poverty rate and select the top 5 community areas
query = """
    SELECT community_area_name, 
           AVG(PERCENT_HOUSEHOLDS_BELOW_POVERTY * 100) AS percent_households_below_poverty
    FROM df
    GROUP BY community_area_name
    ORDER BY percent_households_below_poverty DESC
    LIMIT 5;
"""

# Execute the query
top_poverty_areas = sqldf(query)

# Display the result
print(top_poverty_areas)
```

```python
from pandasql import sqldf
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('ChicagoCrimeData .csv')

# SQL-like query to find the community area with the highest number of crimes
query = """
    SELECT community_area_number, COUNT(*) AS crime_count
    FROM df
    GROUP BY community_area_number
    ORDER BY crime_count DESC
    LIMIT 1;
"""

# Execute the query
most_crime_prone_area = sqldf(query)

# Display the result
print(most_crime_prone_area)
```

```python
from pandasql import sqldf
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('ChicagoCensusData .csv')

# SQL-like query with a subquery to find the name of the community area with the highest hardship index
query = """
    SELECT community_area_name
    FROM df
    WHERE hardship_index = (
        SELECT MAX(hardship_index)
        FROM df
    );
"""

# Execute the query
community_area_with_highest_hardship_index = sqldf(query)

# Display the result
print(community_area_with_highest_hardship_index)
```

```python
from pandasql import sqldf
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('ChicagoCrimeData .csv')

# SQL-like query with a subquery to find the Community Area Name with the most number of crimes
query = """
    SELECT community_area_number
    FROM df
    WHERE iucr = (
        SELECT MAX(iucr)
        FROM df
    );
"""

# Execute the query
community_area_with_most_crimes = sqldf(query)

# Display the result
print(community_area_with_most_crimes)
```
